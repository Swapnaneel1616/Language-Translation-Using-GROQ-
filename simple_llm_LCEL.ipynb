```python
import os
from dotenv import load_dotenv
load_dotenv()

import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
groq_api_key
```

```python
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI
```

```python
model = ChatGroq(model="llama-3.1-8b-instant", api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage , SystemMessage

messages = [
    SystemMessage(content="Translate the following English text to French "),
    HumanMessage(content= "Hello, how are you?" )
]
```

```python
result = model.invoke(messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
#Using LCEL we can chain the components together easily
chain = model|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
generic_template = "Translate the following into {language}:"

prompt = ChatPromptTemplate.from_messages(
    [
        ("system" , generic_template),
        ("user" , "{text}"),
    ]
)

```

```python
result = prompt.invoke({"language": "French" , "text": "Good morning!"})
```

```python
result.to_messages()
```

```python
chain = prompt | model | parser
chain.invoke({"language": "French" , "text": "What is your name?"})
```
